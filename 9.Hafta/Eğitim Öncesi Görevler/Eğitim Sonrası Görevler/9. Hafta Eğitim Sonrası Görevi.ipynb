Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import time
import random


class Kumanda:
    def __init__(self, tv_durumu="turn off", volume=0, channels=["disney","news","TV8"],now_channel= "disney" ):
        self.tv_durumu = tv_durumu
        self.volume = volume
        self.channels = channels
        self.now_channnel = now_channel
    
    def tv_turn_on(self):
        if self.tv_durumu == "Turn On":
            print("TV is already turn on.")
        else:
            print("Tv is turning on...")
            self.tv_durumu = "turn on"


    def tv_turn_off(self):
        if self.tv_durumu == "turn off":
            print("Tv is turn off.")
        else:
            print("Tv is turning off...")
            self.tv_durumu ="turn off"

    def volume_settings(self):
        while True:
            islem = input("Sesi Azalt: '-' /nSrsi Artır:'+'/nÇıkış: 'q'")
            if islem == '-':
                if self.volume >= 0:
                    self.volume -= 1
                    print(f"Volume: {self.volume}")
            elif islem == '+':
                if self.volume != 100:   #max değer 100 olsun
                     self.volume += 1
                     print(f"Volume: {self.volume}")
            elif islem == 'q':
                print("Ses ayarlarından çıkılıyor...")
                break
            else:
                print("Geçersiz işlem.")



    def insert_channel(self,name_of_channel):
        print("Channel is inserting...")
        time.sleep(3)
        self.channels.append(name_of_channel)
        print("The channel is inserted.")

    def learn_channel_num(self):
        return len(self.channels)
    

    def switch_channel(self):
        switch_channel = random.choice(self.channels)
        self.now_channel = switch_channel
        print(f"{self.now_channel} is open now.")

    def TV_informations(self):
        print(f"TV Durumu: {self.tv_durumu}")
        print(f"Volume: {self.volume}")
        print(f"Channels: {self.channels}")
        print(f"Now Channel: {self.now_channel}")
        

def main():
    Kumanda = Kumanda()

print("""
      1. TV turn on
      2. TV turn off
      3. Volume Settings
      4. Insert Channel
      5. Switch Random Channel
      6. Tv Informations
      Click 'q' for out.
      """)



while True:
        islem = input("İşlemi seçiniz: ")
        
        if islem == 'q':
            print("Programdan çıkılıyor...")
            break
        elif islem == '1':
            Kumanda.tv_turn_on()
        elif islem == '2':
            Kumanda.tv_turn_off()
        elif islem == '3':
            Kumanda.volume_settings()
        elif islem == '4':
            name_of_channels = input("Eklemek istediğiniz kanalları ',' ile ayırarak giriniz: ")
            channel = name_of_channels.split(",")
            for channel in channels:
                Kumanda.insert_channel(channel)
        elif islem == '5':
            print(f"Kanal Sayısı: {Kumanda.learn_channel_num()}")
        elif islem == '6':
            Kumanda.switch_channel()
        elif islem == '7':
            Kumanda.TV_informations()
        else:
            print("Geçersiz işlem.")



if __name__ == "__main__":
    main()



        
```
