# Problem 1
`ogrenciler` adında bir sözlük oluşturunuz. Kullanıcıdan döngü yardımıyla 3 öğrenci için veri girmesini isteyerek öğrencilerin sırasıyla okul numarası, isim, soy isim, telefon numarası ve doğum tarihi bilgilerini alınız. Aldığınız öğrenci bilgilerini yine aynı döngü yardımıyla `ogrenciler` adındaki sözlükte saklayınız. Daha sonra öğrenci bilgilerinin bulunduğu sözlük yapısını ekrana bastırınız.



````python

ogrenciler = {}

for i in range(3):
    print(f"{i+1}. Öğrenci bilgilerini giriniz:")
    
    

    okul_numarasi = input("Okul Numarası: ")
    isim = input("İsim: ")
    soy_isim = input("Soy İsim: ")
    telefon = input("Telefon Numarası: ")
    dogum_tarihi = input("Doğum Tarihi: ")

    

    ogrenci_bilgileri = {
        "isim": isim,
        "soy_isim": soy_isim,
        "telefon": telefon,
        "dogum_tarihi": dogum_tarihi
    }


    
    # Öğrenciyi 'okul_numarasi' ile sözlüğe ekliyoruz
    ogrenciler[okul_numarasi] = ogrenci_bilgileri


print("Öğrenci Bilgileri:")
for okul_numarasi, bilgiler in ogrenciler.items():
    print(f"Okul Numarası: {okul_numarasi}")
    print(f"İsim: {bilgiler['isim']}")
    print(f"Soy İsim: {bilgiler['soy_isim']}")
    print(f"Telefon Numarası: {bilgiler['telefon']}")
    print(f"Doğum Tarihi: {bilgiler['dogum_tarihi']}")


#ogrenciler: Bu, anahtar-değer çiftlerini saklayan bir sözlüktür.

#okul_numarasi: Bu, sözlükte bir anahtar olarak kullanılacak olan öğrenciye ait okul numarasıdır. Her öğrenci için benzersiz bir anahtar olması beklenir.

#ogrenci_bilgileri: Bu, okul numarasına karşılık gelen öğrencinin bilgilerini içeren bir sözlüktür. Öğrenci adını, soyadını, telefon numarasını ve doğum tarihini içeren bir sözlüktür.


````

# Problem 2

### a - Biraz ısınarak başlayalım!

```
numbers = [5,10,10,15,15,15,20,20,20,20]
```

Yukarıda `numbers` adında 10 elemanlı bir liste verilmiştir. Bu listenin her elemanını kontrol eden ve tekrar eden elemanları bir kez olacak şekilde başka bir listeye kopyalayan bir program yazınız. Son olarak oluşturduğunuz listeyi ekrana bastırınız.




```
>Output
[5, 10, 15, 20]
```


````python

numbers = [5, 10, 10, 15, 15, 15, 20, 20, 20, 20]

# Tekrar edenleri bir listeye kopyalamak için set kullanıyoruz
unique_numbers = list(set(numbers))
print(unique_numbers)
````

### 🌶️ b) Hadi işi biraz daha zorlaştıralım! 


Şimdiyse senden `numbers` isimli listedeki tekrar eden elemanları `repeatNumbers` adındaki listeye; tekrar etmeyen elemanları `uniqueNumbers`adındaki listeye kopyalayan yazmanı ve listeleri ekrana bastırmanı bekliyoruz

```
>Output
Tekrar Eden Elemanlar: [10, 10, 15, 15, 15, 20, 20, 20, 20]
Tekrar Etmeyen Elemanlar: [5]
```

````python

numbers = [5, 10, 10, 15, 15, 15, 20, 20, 20, 20]

# boş listeler oluşturduk
repeat_numbers = []
unique_numbers = []

# Tekrar sayılarını takip etmek için bir sözlük oluşturuyoruz
count_dict = {}

# Her elemanın tekrar sayısını hesaplamak için
for num in numbers:
    if num in count_dict:
        count_dict[num] += 1
    else:
        count_dict[num] = 1

# Tekrar eden ve tekrar etmeyen elemanları ayırdık
for num in numbers:
    if count_dict[num] > 1:
        repeat_numbers.append(num)
    elif count_dict[num] == 1:
        unique_numbers.append(num)


print(repeat_numbers)
print( unique_numbers)
````

# Problem 4

Buradaki işlemlerimizin tamamı stringler içindir. Aşağıda verilecek olan metodlar için kendinizin örnek yaratması gerekmektedir. 

Örneğin `upper` için `string` değişkeni oluştururken` k = "BAC"` şeklinde oluşturmayınız. Bunun yerine k değişkenini `k = "bac`" şeklinde oluşturmanız `upper` metodundaki değişimi görebilmeniz için daha iyi olacaktır.

Bütün metodlardaki değişimin gözle görülür şekilde değişimi beklenmektedir.


1- capitalize(), lower(), upper(), swapcase(), title(), replace()

2- endswith(), startswith(), find(), index(), count()

3- split(), strip(), join(), islower(), isalpha(), isdigit()


Yukarıdaki `string` metodlarındaki değişimleri görebildiğimiz bir program yazınız.


````python

string = "   hi i am Percy! hold fast mom.   "



# capitalize() İlk harfi büyük yapar
print(string)
print( string.capitalize())
print()

# lower()  Tüm harfleri küçük yapar
print(string)
print(string.lower())
print()

# upper()  Tüm harfleri büyük yapar
print(string)
print(string.upper())
print()

# swapcase()  Büyük harfleri küçük, küçük harfleri büyük yapar
print(string)
print( string.swapcase())
print()

# title()  Her kelimenin ilk harfini büyük yapar
print(string)
print( string.title())
print()

# replace() 
print(string)
print( string.replace("Percy", "Jackson"))
print()



# endswith()  String'in belirli bir suffix ile bitip bitmediğini kontrol eder
print(string)
print(string.endswith("mom."))
print()

# startswith()  String'in belirli bir prefix ile başlayıp başlamadığını kontrol eder
print(string)
print(string.startswith("   hi"))
print()

# find() Belirli bir alt string'in ilk bulunma indeksini döner, bulunamazsa -1 döner
print(string)
print(string.find("Percy"))
print()

# index()  Belirli bir alt string'in ilk bulunma indeksini döner, bulunamazsa hata verir
print(string)
print(string.index("hold"))
print()


# count()  Belirli bir alt string'in kaç kez geçtiğini döner
print(string)
print( string.count("mom"))
print()


# split() String'i belirli bir separator ile böler
print(string)
print(string.split())
print()

# strip()  Baş ve sondaki boşlukları temizler
print(string)
print( string.strip())
print()

# join()  Bir listeyi bir separator ile birleştirir
print(string)
words = string.split()
print( ' '.join(words))
print()

# islower() - Tüm karakterlerin küçük olup olmadığını kontrol eder
print(string)
print(string.islower())
print()

# isalpha() - String'in sadece alfabetik karakterler içerip içermediğini kontrol eder
print(string)
print("hold".isalpha())
print("i am Percy".isalpha())
print()

# isdigit() - String'in sadece rakamlar içerip içermediğini kontrol eder
print(string)
print( "123".isdigit())
print("123abc".isdigit())
print()
````
