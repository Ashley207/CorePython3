# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
# 2.Problemin Çözümünü Buraya Yazınız

def kelime_degistir(dosya_adi, yeni_dosya_adi, eski_kelime, yeni_kelime):
    try:
        with open(dosya_adi, "r") as dosya:
            icerik = dosya.read()

    except FileNotFoundError:
        print(f"{dosya_adi} dosyası bulunamadı.")
        return
    
    
    yeni_icerik = icerik.replace(eski_kelime, yeni_kelime)
    
    try:
        with open(yeni_dosya_adi, "w") as yeni_dosya:
            yeni_dosya.write(yeni_icerik)
        print("Kelime başarıyla değiştirildi ve yeni dosyaya yazıldı.")
    except IOError:
        print(f"{yeni_dosya_adi} dosyasına yazma hatası.")


eski_kelime = input("Değiştirmek istediğiniz kelime: ")
yeni_kelime = input("Yerine geçecek yeni kelime: ")


kelime_degistir("urunler.txt", "urunler_yeni.txt", eski_kelime, yeni_kelime)



````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
# self, yöntemlerin sınıfın örneğiyle ilişkilendirilmesini sağlar.
#  __init__ (başlatıcı) yöntemini çağırmasını sağlar. Bu, alt sınıfın üst sınıfın özelliklerini ve yöntemlerini miras almasını ve başlatmasını mümkün kılar.
# Bus sınıfı, Vehicle sınıfından miras alır ve __init__ yönteminde super().__init__(name, speed, mileage) ifadesi ile Vehicle sınıfının __init__ yöntemini çağırır.
#super() fonksiyonu, üst sınıfın bir örneğini döndürür ve bu örnek üzerinden üst sınıfın yöntemleri çağrılabilir. Bu da kod tekrarını azaltır.




class Vehicle:
    def __init__(self, name, speed, mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage

    def display_info(self):
        print(f"Name: {self.name} Speed: {self.speed} Mileage: {self.mileage}")

class Bus(Vehicle):
    def __init__(self, name, speed, mileage):
        super().__init__(name, speed, mileage)


school_bus = Bus("School Volvo", 180, 12)
school_bus.display_info()
````
